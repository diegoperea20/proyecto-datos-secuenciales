## Transcripción de voz 🗣️

Ahora, vamos a transcribir tu archivo de audio. Puedes ajustar la `batch_size` y el `compute_type` según tus necesidades.

```python
import whisper
import gc

# Transcripción
model = whisper.load_model("medium")
audio = whisper.load_audio(audio_file)
result = model.transcribe(audio)
print(result["segments"])
print(f'resultado:{result}')

# Alineación
for segment in result["segments"]:
    start_time = segment["start"]
    end_time = segment["end"]
    transcript = segment["text"]
    print(f'Tiempo de inicio: {start_time}, Tiempo de fin: {end_time}, Transcripción: {transcript}')
```

⬇️ Si quieres **descargar la transcripción** como un archivo de texto, ejecuta la siguiente celda.

```python
# Concatenating the transcribed segments
transcription_text = "\n".join([segment['text'] for segment in result["segments"]])

# Writing to a .txt file
with open('transcription.txt', 'w') as file:
    file.write(transcription_text)

# Download the .txt file
#files.download('transcription.txt')
```

⬇️ Y si quieres **en .json la transcripción al completo** junto a sus timestamps, ejecuta la siguiente celda.

```python
import json

# Saving the entire result as a JSON file
with open('transcription.json', 'w') as file:
    json.dump(result, file, indent=4)

# Download the JSON file
#files.download('transcription.json')
```

```python
from transformers import pipeline
import json

# Carga el modelo de traducción de español a inglés
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-es-en")

# Carga el archivo JSON original
with open("/content/transcription.json", "r") as f:
    data = json.load(f)

# Traduce el texto principal
data["text"] = translator(data["text"], max_length=40)[0]["translation_text"]

# Recorre cada segmento para traducir el texto
for segment in data["segments"]:
    segment["text"] = translator(segment["text"], max_length=40)[0]["translation_text"]

# Escribe el diccionario en el archivo JSON original
with open("archivoEnglish.json", "w") as f:
    json.dump(data, f, indent=4)

#files.download('archivoEnglish.json')
```

```python
# Concatenando los segmentos traducidos
transcription_text_english = "\n".join([segment['text'] for segment in data["segments"]])

# Escribiendo en un archivo .txt
with open('transcriptionEnglish.txt', 'w') as file:
    file.write(transcription_text_english)

#files.download('transcriptionEnglish.txt')
```
